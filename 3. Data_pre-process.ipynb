# 1 INITIAL STEPS

We will load the necessary libraries and define functions to convert the initial data into dataframes.

## Import libraries

```python
import arff
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

## Function definition

```python
def load_dataset(data_path):
    with open(data_path, 'r') as train_set:
        dataset = arff.load(train_set)
    attributes = [attr[0] for attr in dataset["attributes"]]
    return pd.DataFrame(dataset["data"], columns=attributes)
```

```python
def sets_split(df, rstate=42, shuffle=True, stratify=None):
    strat = df[stratify] if stratify else None
    train_set, test_set = train_test_split(
        df, test_size=0.4, random_state=rstate, shuffle=shuffle, stratify=strat)
    strat = test_set[stratify] if stratify else None
    val_set, test_set = train_test_split(
        test_set, test_size=0.5, random_state=rstate, shuffle=shuffle, stratify=strat)
    return (train_set, val_set, test_set)
```

```python
ruta="../../datasets/NSL-KDD/KDDTrain+.arff" #note tha ruta means path, check files are located in this path
```

# 2 INITIAL DATA TRERATMENT

```python
df_raw = load_dataset(ruta) #dataset loading
```

```python
df_raw
```

```python
#Data stratification using protocol_type as point of stratification
train_set, val_set, test_set = sets_split(df_raw, stratify='protocol_type')
```

```python
print("Training set size", len(train_set))
print("Validation set size:", len(val_set))
print("Test set size:", len(test_set))
```

# 3 CLEANING DATA

We split input and output data for the training test

Then we treat N.A values

```python
# Splitting input and outup values
X_train = train_set.drop("class", axis=1)
y_train = train_set["class"].copy()
```

```python
# Creating som N.A values
X_train.loc[(X_train["src_bytes"]>500) & (X_train["src_bytes"]<700), "src_bytes"] = np.nan
X_train.loc[(X_train["dst_bytes"]>600) & (X_train["dst_bytes"]<2500), "dst_bytes"] = np.nan
X_train
```

## 3.1 Removinf NA rows

```python
X_train.isna().any()
```

```python
# Selection fo the NA rows
filas_valores_nulos  = X_train[X_train.isnull().any(axis=1)]
filas_valores_nulos
```

```python
# Copy of the dataset in order to do NOT change it
X_train_copy = X_train.copy()
```

```python
# Removing NA rows
X_train_copy.dropna(subset=["src_bytes", "dst_bytes"], inplace=True)
X_train_copy
```

```python
print("# Rows removed:", len(X_train) - len(X_train_copy))
```

## 3.2 Removing NA featurs

```python
# Copy of the dataset
X_train_copy = X_train.copy()
```

```python
# Removing features
X_train_copy.drop(["src_bytes", "dst_bytes"], axis=1, inplace=True)
X_train_copy
```

```python
print("# Features removed:", len(list(X_train)) - len(list(X_train_copy)))
```

## 3.3 Replcament of NA values with another value

```python
# Copy of the dataset
X_train_copy = X_train.copy()
```

```python
# Using average value as replace N.A value
media_srcbytes = X_train_copy["src_bytes"].mean()
media_dstbytes = X_train_copy["dst_bytes"].mean()

X_train_copy["src_bytes"] = X_train_copy["src_bytes"].fillna(media_srcbytes)
X_train_copy["dst_bytes"] = X_train_copy["dst_bytes"].fillna(media_dstbytes)

X_train_copy
```
